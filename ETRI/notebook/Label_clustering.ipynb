### 이 코드는 ETRI의 형태소 분석 API를 적용한 코드로, 앞서 국제표준산업분류 데이터에서 공백 기준으로 텍스트 데이터를 ETRI 형태소 분석 API를 통해 명사만 추출하여 저장한 코드입니다.

```python
#-*- coding:utf-8 -*-
import urllib3
import json
from tqdm import tqdm

openApiURL = "http://aiopen.etri.re.kr:8000/WiseNLU" 

accessKey = "692fa2ee-d55d-4581-95c7-7c094be35280"
analysisCode = "morp"

text = ''
total_nouns_lst = []
for i in tqdm(list(set(total_lst))):
    if len(text) <= 200:
      text += i + ' '
    else:
      requestJson = {
          "access_key": accessKey,
          "argument": {
              "text": text,
              "analysis_code": analysisCode
          }
      }

      http = urllib3.PoolManager()
      response = http.request(
          "POST",
          openApiURL,
          headers={"Content-Type": "application/json; charset=UTF-8"},
          body=json.dumps(requestJson)
      )
      
      text = ''
      total_nouns_lst += json.loads(str(response.data,"utf-8"))['return_object']['sentence'][0]['morp']
      
total_nouns_lst
```

## 6. total_data.csv의 description에서 명사만 추출된 데이터와 국제표준산업분류에서의 명사만 추출된 데이터를 교집합 해서 `final_nouns` 데이터 만들기

```python
total_df['final_nouns'] = ''
for index, row in tqdm(total_df.iterrows()):
  try: total_df.loc[index, 'final_nouns'] = list(set(total_df.loc[index, 'nouns']) & set(final_noun_lst))
  except: pass

total_df
```

```python
total_df['final_nouns']
```

```python
final_nouns_lst = []
for i in tqdm(total_df['final_nouns']):
  data = list(set(i))
  final_nouns_lst.append(' '.join(data))

final_nouns_lst
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorize = CountVectorizer()
x = vectorize.fit_transform(final_nouns_lst)
```

## 7. K-means 알고리즘을 통해 K = 15 즉, 15개의 군집 생성해서 결과 확인

```python
from sklearn.cluster import KMeans

######### 평균 군집 분석 ###########
#데이터 프레임 생성 (문장 단어 매트릭스 만들기)
new_df = pd.DataFrame(x.toarray(), columns=vectorize.get_feature_names_out())

# K- 평균 군집 분석
kmeans = KMeans(n_clusters=15).fit(new_df)
print('[K-평균 군집 분석 결과]')
print('###########################################')

print(kmeans.labels_)
```

```python
label = list(kmeans.labels_)
total_df['kmeans_label'] = label
total_df
```

```python
from pyclustering.cluster import kmedoids

######### k-대푯값 군집 분석 ###########
#인스턴스 생성
kmedoids_ins = kmedoids.kmedoids(new_df.values,initial_index_medoids=[4,18,36]) # initial_index는 df행 중 대푯값 행 지정
kmedoids_ins.process()
clusters = kmedoids_ins.get_clusters()
print('###########################################')
print('[K-대푯값 군집 분석 결과]')

print(clusters)
```
