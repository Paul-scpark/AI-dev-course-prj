```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
full_df = pd.read_csv("./total_data.csv")
```

```python
full_df
```

```python
full_x = full_df.drop("label", axis=1)
full_y = full_df["label"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(full_x, full_y, test_size=0.1, random_state=42, stratify=full_y)
```

```python
x_train["label"] = y_train
x_test["label"] = y_test
```

```python
x_train.to_csv("./train.csv", index=None)
x_test.to_csv("./test.csv", index=None)
```
