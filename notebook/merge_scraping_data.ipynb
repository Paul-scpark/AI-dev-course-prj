## 1. 공공데이터포털 데이터 취합

```python
data_go = pd.read_csv('data_go_kr_data.csv', index_col=0)
print(data_go.shape)
data_go.head(1)
```

```python
data_go = data_go[['서비스', '설명', 'url', '매체유형', '제공기관', '분류체계']]
data_go['ori_source'] = '공공데이터포털'
data_go.columns = ['title', 'description', 'url', 'data_type', 'source', 'ori_label', 'ori_source']

data_go['ori_label'] = data_go['ori_label'].str.extract('([\S]+)').iloc[:, 0]
data_go['label'] = data_go.replace({'ori_label' : {
    '산업·통상·중소기업': '제조소비',
    '농림': '식품건강',
    '보건': '보건의료',
    '재정·세제·금융': '경제금융',
    '일반공공행정': '행정법률',
    '사회복지': '사회복지',
    '환경': '환경기상',
    '교통및물류': '교통물류',
    '문화체육관광': '문화관광',
    '해양수산': '식품건강',
    '공공질서및안전': '재난안전',
    '지역개발': '주택토지',
    '통신': '교육과학',
    '교육': '교육과학',
    '과학기술': '교육과학',
    '통일·외교': '통일외교안보',
    '국방': '통일외교안보'
}})['ori_label']
data_go
```

## 2. 빅데이터 지도 데이터 취합

```python
bigdata_map = pd.read_csv('bigdata_map_data.csv')
print(bigdata_map.shape)
bigdata_map.head(1)
```

```python
bigdata_map = bigdata_map[['title', 'desc', 'landing_page', 'file_extensions', 'platform_title', 'large_category']]
bigdata_map['ori_source'] = '빅데이터지도'
bigdata_map['file_extensions'] = bigdata_map['file_extensions'].str.replace('[', '').str.replace(']', '').str.replace("'", '')
bigdata_map.columns = ['title', 'description', 'url', 'data_type', 'source', 'ori_label', 'ori_source']

bigdata_map['label'] = bigdata_map.replace({'ori_label' : {
    '문화체육': '문화관광',
    '교육과학': '교육과학',
    '교통통신': '교통물류',
    '경제금융': '경제금융',
    '건설에너지': '주택토지',
    '농수산식품': '식품건강',
    '제조소비': '제조소비',
    '재난안전': '재난안전',
    '환경자원': '환경기상',
    '행정법률': '행정법률'
}})['ori_label']
bigdata_map
```

## 3. AI-hub 데이터 취합

```python
ai_hub = pd.read_csv('ai_hub_data.csv', index_col=0)
print(ai_hub.shape)
ai_hub.head(1)
```

```python
ai_hub = ai_hub[['title', '소개', 'url', '데이터 유형', '데이터 출처', '데이터 영역']]
ai_hub['ori_source'] = 'AI_Hub'
ai_hub.columns = ['title', 'description', 'url', 'data_type', 'source', 'ori_label', 'ori_source']

ai_hub['label'] = ai_hub.replace({'ori_label' : {
    '교통물류 ': '교통물류',
    '한국어 ': '기타',
    '농축수산 ': '식품건강',
    '영상이미지 ': '기타',
    '헬스케어 ': '보건의료',
    '재난안전환경 ': '재난안전',
    '데이터 영역': '기타',
    '텍스트': '기타',
    '이미지': '기타',
    '오디오 , 비디오': '기타',
    '오디오 , 텍스트': '기타'
}})['ori_label']
ai_hub
```

## 4. 서울 열린데이터 광장 데이터 취합

```python
data_seoul = pd.read_csv('data_seoul.csv', index_col=0)
print(data_seoul.shape)
data_seoul['source'] = data_seoul['제공기관'] + ' ' + data_seoul['제공부서']
data_seoul.head(1)
```

```python
data_seoul = data_seoul[['이름', '명세', 'URL', '제공형식', 'source', '카테고리']]
data_seoul['ori_source'] = '서울데이터광장'
data_seoul.columns = ['title', 'description', 'url', 'data_type', 'source', 'ori_label', 'ori_source']

data_seoul['label'] = data_seoul.replace({'ori_label' : {
    '교통': '교통물류',
    '복지': '사회복지',
    '주택/건설': '주택토지',
    '도시관리': '주택토지',
    '환경': '환경기상',
    '일반행정': '행정법률',
    '안전': '재난안전',
    '보건': '보건의료',
    '문화/관광': '문화관광',
    '산업/경제': '경제금융',
    '교육': '교육과학'
}})['ori_label']
data_seoul
```

## 5. 취합한 데이터를 하나로 만들기

```python
total_df = pd.concat([data_go, bigdata_map, ai_hub, data_seoul], axis=0).reset_index(drop=True)
total_df
```

### 팀에서 구분해놓은 13개 + 1개 (기타) = 14개 Label

제조소비, 식품건강, 보건의료, 경제금융, 행정법률, <br/>
사회복지, 환경기상, 교통물류, 문화관광, 재난안전, <br/>
주택토지, 교육과학, 통일외교안보, 기타 (한국어, 이미지, 텍스트 등)

```python
total_df['label'].unique()
```

### 일부 데이터는 출처 (URL)가 겹치는 항목들도 존재
URL 값을 기준으로 unique한 것만 찾으면, 기존에 127,051개에서 94,805개로 변화

```python
len(total_df['url'].unique())
```

### URL이 겹치는 항목들에서 title, ori_label, label 데이터 확인
확인해보니, 데이터 제공처마다 유사한 데이터지만, 세부적으로는 label이 다른 경우들도 존재 (우리 label 기준도 다르게 나옴)

```python
for i, _ in total_df.value_counts('url').head(30)[1:].iteritems():
    print(total_df[total_df['url'] == i][['title', 'ori_label', 'label']], '\n')
```
